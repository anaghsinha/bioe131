```python
import re, string 
import matplotlib.pyplot as plt
import numpy as np

rjfile = open("romeoandjuliet.txt")
rjstring = rjfile.read()

exclude = set(string.punctuation)

def remove_punctuation(s):
    return ''.join(ch for ch in s if ch not in exclude)

rjstring = remove_punctuation(rjstring)
rjstring = rjstring.lower()


print (rjstring)
```

```python
rjdict = dict()
```

```python
for word in rjstring.split():
    if word in rjdict:
        rjdict[word] += 1
    else:
        rjdict[word] = 1


rjdict_sorted = sorted(rjdict, key = rjdict.get, reverse = True)

rjdict_sorted_first = rjdict_sorted[:10]
rjdict_sorted_last = rjdict_sorted[-10:]

print(rjdict_sorted_first)
print(rjdict_sorted_last)
```

```python
plt.bar(height=[rjdict[word] for word in rjdict_sorted_first],x=rjdict_sorted_first)

plt.show()
```

```python
plt.figure(figsize=(10, 5))  # width:20, height:3
plt.bar(x=range(len(rjdict_sorted_last)), height=[rjdict[word] for word in rjdict_sorted_last], align='center')


print(len(rjdict_sorted_last))
plt.xticks(range(len(rjdict_sorted_last)), rjdict_sorted_last)

plt.show()
```
