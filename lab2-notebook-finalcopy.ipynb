```python
#how to make GC content analyser function
#create a list with 2238 0's (must we adjust for diff seq lens? not so sure)
#take any one sequence, admit only the sequences in our cluster
#if at position 0, they have G/C, add 1 to that position
#cycle through each of the other sequences
#divide each sequence by the length of the cluster file to create GC content
```

```python
def gc_counting(clusternum):
    gc_counter = []
    for posn in range(2238):
        gc_counter.append(0) #create a list with 2238 0's
    
    for record in align: #take any one sequence, admit only the sequences in our cluster
        if record.id in clusternum:
            for posn in range(2238):
                if record.seq[posn] == 'G' or record.seq[posn] == "C": #if at position 0, they have G/C, add 1 to that position
                    gc_counter[posn] += 1
        #cycle through each of the other sequences
        
    return [freq/len(clusternum) for freq in gc_counter] #divide each sequence by the length of the cluster list to create GC content
```

```python
'''Creating an at_content analyser function works the same way as gc_counting, detailed above'''
def at_counting(clusternum):
    at_counter = []
    for posn in range(2238):
        at_counter.append(0) #create a list with 2238 0's
    
    for record in align: #take any one sequence, admit only the sequences in our cluster
        if record.id in clusternum:
            for posn in range(2238):
                if record.seq[posn] == 'A' or record.seq[posn] == "T": #if at position 0, they have G/C, add 1 to that position
                    at_counter[posn] += 1
        #cycle through each of the other sequences
        
    return [freq/len(clusternum) for freq in at_counter] #divide each sequence by the length of the cluster list to create GC content

#print(at_counting(cluster1)) #testing at_counting
```

```python
#generating boxplot with first 100 bases, all 2000+ bases is shown underneath
import numpy as np
import matplotlib.pyplot as plt

for i in range(1,8):
    
    N = 100
    gc = gc_counting(eval("cluster" + str(i)))[:N]
    at = at_counting(eval("cluster" + str(i)))[:N]
    ind = np.arange(N)    # the x locations for the groups
    # the width of the bars is left at default: can also be len(x) sequence

    plt.figure(figsize=(18,6))
    p1 = plt.bar(ind, gc)
    p2 = plt.bar(ind, at, 
                 bottom=gc)

    plt.ylabel('GC/AT scores')
    plt.title('GC scores of each position of the alignment in cluster %d' %i)
    plt.yticks(np.arange(0, 1.5, .1))
    plt.legend((p1[0], p2[0]), ('G/C', 'A/T'))

    plt.show()
```

```python
#it is definitely possible to use the unaligned sequences in seqs.fa to easily come up with a sequence length for each sequence.
#however, given that the lab2 pdf mentions that "youll need to have loaded the seqs.aligned.fa file you obtained above
#into your iPython notebook using Bio.SeqIO" to complete this part - I'm going to write code with this file, even if it seems a little needless

def len_counting(clusternumber):
    len_counter = []
    
    for record in align: #take any one sequence, admit only the sequences in our cluster
        count_alphabets = 0
        if record.id in clusternumber:
                for alphabet in record.seq:
                    if alphabet != '-':
                        count_alphabets += 1
                len_counter.append(count_alphabets)
        #cycle through each of the other sequences
        
    return len_counter #divide each sequence by the length of the cluster list to create GC content
```

```python
length_store = []
for i in range(1,8):
    length_store.append((len_counting(eval("cluster" + str(i)))))
    
print(length_store)
```
