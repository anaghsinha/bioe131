The Compression Efficiency deacreases as the number of percentage of zeroes decreases to 50 percent because of the increasing entropy of the sequence as you go away from 100% zeroes. Increasing the entropy of a sequence increases its Shannon information, and as a result it takes more bits to encode the sequence by a compression algorithm than it would have otherwise.

The minimum number of bits required to store a single DNA base (assuming you only need to encode the 4 bases, and assuming the code is not optimized for any sequence other than a randomly generated string) is log(base 2) 4 = 2 bits

The minimum number of bits required to store a single amino acid (assuming you only need to encode the 20 amino, and assuming the code is not optimized for any sequence other than a randomly generated string) is ceiling(log(base 2) 20) = ceiling(4.32) = 5 bits. 

gzip required 484,471,080 bits to store 100 million letters of DNA bases @ 4.84 bits per base
bzip2 required 442,055,872 bytes to store 100 million letters of DNA bases @ 4.42 bits per base

gzip required 838,994,024 bits to store 100 million letters of AA residues @ 8.38 bits per residue
bzip2 required 442,055,872 bits to store 100 million letters of AA residues @ 4.42 bits per residue

Both gzip and bzip perform rather poorly on the DNA sequence files, as they both achieve about double the theoretical limit for an ideal compression algorithm that should code for only 2 bases. Nonetheless, it is still smaller than the starting file by a factor of 2 so it's fairly decent.

gzip however does terribly on the amino acid sequence, having increased the size of the AA file slightly from the uncompressed version. However, bzip2 does a good job of compression on this file, actually managing to come very close to the true theoretical Shannon limit of 4.32 bits per residue.
